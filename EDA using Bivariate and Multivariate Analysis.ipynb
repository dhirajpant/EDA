```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading the Titanic dataset into a DataFrame called 'titanic
titanic = pd.read_csv('train.csv')
```

```python
titanic.head()
```

## 1.Scatterplot(NUM-NUM)

```python
# Set up the plotting environment
plt.figure(figsize=(10, 6))

# Create a scatter plot
sns.scatterplot(data=titanic, x='Age', y='Fare')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')

# Display the plot
plt.show()
```

```python
# Set up the plotting environment
plt.figure(figsize=(10, 6))

# Create a scatter plot
sns.scatterplot(data=titanic, x='Age', y='Fare',hue='Sex')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')

# Display the plot
plt.show()
```

```python
# Set up the plotting environment
plt.figure(figsize=(10, 6))

# Create a scatter plot
sns.scatterplot(data=titanic, x='Age', y='Fare',hue='Sex',style='Pclass')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')

# Display the plot
plt.show()
```

```python
# Set up the plotting environment
plt.figure(figsize=(10, 6))

# Create a scatter plot
sns.scatterplot(data=titanic, x='Age', y='Fare',hue='Sex',style='Pclass',size='Survived')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')

# Display the plot
plt.show()
```

## 2.Bar Plot(NUM-CATE)

```python
titanic.head()
```

```python
# Set up the plotting environment
plt.figure(figsize=(10, 6))

# Create a bar plot
sns.barplot(data=titanic, x='Pclass', y='Age')

# Add labels and title
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.title('Bar Plot of Age by Passenger Class')

# Display the plot
plt.show()
```

```python
# Set up the plotting environment
plt.figure(figsize=(10, 6))

# Create a bar plot
sns.barplot(data=titanic, x='Pclass', y='Fare',hue='Sex')

# Add labels and title
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.title('Bar Plot of Fare by Passenger Class')

# Display the plot
plt.show()
```

## 3.Box Plot(NUM-CATE)

```python
# Set up the plotting environment
plt.figure(figsize=(10, 6))

# Create a box plot
sns.boxplot(data=titanic, x='Sex', y='Age')

# Add labels and title
plt.xlabel('Sex')
plt.ylabel('Age')
plt.title('Box Plot of Age by Sex')

# Display the plot
plt.show()
```

```python
# Set up the plotting environment
plt.figure(figsize=(10, 6))

# Create a box plot
sns.boxplot(data=titanic, x='Sex', y='Age',hue='Survived')

# Add labels and title
plt.xlabel('Sex')
plt.ylabel('Age')
plt.title('Box Plot of Age by Sex')

# Display the plot
plt.show()
```

## 4.Histplot(NUM-CATE)

```python
# Set up the plotting environment
plt.figure(figsize=(10, 6))

# Create a histogram
sns.histplot(data=titanic, x='Age', hue='Survived', multiple='stack')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age Segmented by Survival')

# Display the plot
plt.show()
```
